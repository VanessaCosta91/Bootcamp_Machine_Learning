```python
from google.colab import files
from PIL import Image
```

```python
uploaded = files.upload()
```

```python
imagem = Image.open('paisagem.jpg')
pixels = imagem.load()
largura, altura = imagem.size
```

```python
imagem_cinza = Image.new('L', (largura, altura))

for x in range(largura):
  for y in range(altura):
    r, g, b = pixels[x, y]
    cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
    imagem_cinza.putpixel((x, y), cinza)
```

```python
imagem_pb = Image.new('1', (largura, altura))

limite = 127

for x in range(largura):
  for y in range(altura):
    valor_cinza = imagem_cinza.getpixel((x, y))
    pb = 255 if valor_cinza >= limite else 0
    imagem_pb.putpixel((x, y), pb)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(imagem_cinza, cmap='gray')
plt.axis('off')
plt.title('Imagem em tons de cinza')
plt.show()
```

```python
plt.imshow(imagem_pb, cmap='gray')
plt.axis('off')
plt.title('Imagem em preto e branco')
plt.show()
```
